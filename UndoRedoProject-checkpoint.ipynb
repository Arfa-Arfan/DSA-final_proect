# Undo/Redo — Project
This notebook provides a text Undo/Redo implementation with a multi-document CLI that avoids manual index handling.

How to run:
- Run Cell 1 to define the class (`UndoRedo`).
- Run Cell 3 to define the editor and commands (`NotebookEditor`, `run_command`).
- Run Cell 5 to define the interactive loop (`start_cli`).
- Run Cell 6 to start the prompt.

Key commands (no indexes needed):
- `new <name>`, `use <name>`, `list`
- `write <text>`, `set <text>`, `clear`, `backspace [n]`, `delete [n]`
- `show`, `undo`, `redo`, `replaceall <old> <new>`, `next`, `help`, `exit`
- Advanced (optional): `insertat <pos> <text>`, `replaceat <pos> <len> <text>`

```python
from __future__ import annotations
from typing import List

class UndoRedo:
    """
    Text undo/redo using two stacks.
    Maintains prior states in `undo_stack` and forward states in `redo_stack`
    following a standard LIFO (last-in, first-out) strategy.
    """

    def __init__(self) -> None:
        # Two stacks to store previous and redo states
        self.undo_stack: List[str] = []
        self.redo_stack: List[str] = []
        # Current text (starts empty)
        self._text = ""

    @property
    def text(self) -> str:
        # Return the current text
        return self._text

    def next_index(self) -> int:
        # Index where the next character would be appended
        return len(self._text)

    # --- Editing operations (each saves the pre-edit state) ---
    def insert(self, pos: int, value: str) -> None:
        # Validate position (end-inclusive is allowed)
        self._validate_pos_insert(pos)
        before = self._text
        # Insert value at position
        self._text = self._text[:pos] + value + self._text[pos:]
        self._push_undo(before)

    def delete(self, pos: int, length: int) -> None:
        self._validate_range(pos, length)
        before = self._text
        end = pos + length
        # Remove slice [pos:end)
        self._text = self._text[:pos] + self._text[end:]
        self._push_undo(before)

    def replace(self, pos: int, length: int, value: str) -> None:
        self._validate_range(pos, length)
        before = self._text
        end = pos + length
        # Replace slice with new value
        self._text = self._text[:pos] + value + self._text[end:]
        self._push_undo(before)

    # --- Undo / Redo ---
    def undo(self) -> bool:
        # Return False when there is nothing to undo
        if not self.undo_stack:
            return False
        prev = self.undo_stack.pop()
        self.redo_stack.append(self._text)
        self._text = prev
        return True

    def redo(self) -> bool:
        # Return False when there is nothing to redo
        if not self.redo_stack:
            return False
        nxt = self.redo_stack.pop()
        self.undo_stack.append(self._text)
        self._text = nxt
        return True

    # --- Helpers ---
    def _push_undo(self, before_state: str) -> None:
        # Push the pre-edit state and invalidate redo on new edits
        self.undo_stack.append(before_state)
        self.redo_stack.clear()

    def _validate_pos_insert(self, pos: int) -> None:
        if pos < 0 or pos > len(self._text):
            raise IndexError("insert position out of range")

    def _validate_range(self, pos: int, length: int) -> None:
        if length < 0:
            raise ValueError("length must be non-negative")
        if pos < 0 or pos + length > len(self._text):
            raise IndexError("range out of range")
```

```python
# Command helpers and multi-document editor
import re
from typing import List, Optional, Dict

class NotebookEditor:
    """
    Manages multiple named documents backed by `UndoRedo`.
    Provides index-free operations like `write`, `backspace`, and `set`.
    """
    def __init__(self) -> None:
        self.docs: Dict[str, UndoRedo] = {}
        self.current: Optional[str] = None

    def new(self, name: str) -> str:
        self.docs[name] = UndoRedo()
        self.current = name
        return f"created and using: {name}"

    def use(self, name: str) -> str:
        if name not in self.docs:
            self.docs[name] = UndoRedo()
        self.current = name
        return f"using: {name}"

    def list_docs(self) -> str:
        if not self.docs:
            return "(no documents)"
        hint = " (current)"
        return "\n".join(f"- {n}{(hint if n == self.current else '')}" for n in self.docs)

    def _cur(self) -> UndoRedo:
        if self.current is None:
            # create a default doc on first use
            self.new("default")
        return self.docs[self.current]  # type: ignore[index]

    def show(self) -> str:
        return self._cur().text

    def next_index(self) -> int:
        return self._cur().next_index()

    def write(self, text: str) -> str:
        doc = self._cur()
        doc.insert(doc.next_index(), text)
        return doc.text

    def set(self, text: str) -> str:
        doc = self._cur()
        doc.replace(0, len(doc.text), text)
        return doc.text

    def clear(self) -> str:
        return self.set("")

    def backspace(self, n: int = 1) -> str:
        doc = self._cur()
        n = max(0, min(n, len(doc.text)))
        if n:
            doc.delete(len(doc.text) - n, n)
        return doc.text

    def delete(self, n: int = 1) -> str:
        return self.backspace(n)

    def insert_at(self, pos: int, text: str) -> str:
        doc = self._cur()
        doc.insert(pos, text)
        return doc.text

    def replace_range(self, pos: int, ln: int, text: str) -> str:
        doc = self._cur()
        doc.replace(pos, ln, text)
        return doc.text

    def replace_all(self, old: str, new: str) -> str:
        doc = self._cur()
        updated = doc.text.replace(old, new)
        doc.replace(0, len(doc.text), updated)
        return doc.text

    def undo(self) -> str:
        doc = self._cur()
        return doc.text if doc.undo() else "Nothing to undo"

    def redo(self) -> str:
        doc = self._cur()
        return doc.text if doc.redo() else "Nothing to redo"

HELP = """
Undo/Redo (index-free defaults)
Commands:
  new <name>            # create and switch to a new document
  use <name>            # switch to document (creates if missing)
  list                  # list documents, marking current
  write <text>          # append text to the end
  set <text>            # replace entire content
  clear                 # erase entire content
  backspace [n]         # delete last n chars (default 1)
  delete [n]            # alias of backspace
  insertat <pos> <text> # insert at index (advanced)
  replaceat <pos> <len> <text>  # range replace (advanced)
  replaceall <old> <new> # replace all occurrences
  next                  # show next write index
  show                  # print current content
  undo                  # undo last change
  redo                  # redo last undo
  help                  # show this help
  exit                  # quit loop
"""

def run_command(raw: str, ed: NotebookEditor) -> str:
    """Parse and execute a command against a `NotebookEditor`."""
    if not raw or not raw.strip():
        return ""
    parts = raw.split()
    cmd = parts[0].lower()

    try:
        if cmd == "help":
            return HELP.strip()
        elif cmd == "list":
            return ed.list_docs()
        elif cmd == "new":
            if len(parts) != 2:
                return "Usage: new <name>"
            return ed.new(parts[1])
        elif cmd == "use":
            if len(parts) != 2:
                return "Usage: use <name>"
            return ed.use(parts[1])
        elif cmd == "show":
            return ed.show()
        elif cmd == "next":
            return str(ed.next_index())
        elif cmd == "write":
            m = re.match(r"^\s*write\s(.*)$", raw)
            if not m:
                return "Usage: write <text>"
            return ed.write(m.group(1))
        elif cmd == "set":
            m = re.match(r"^\s*set\s(.*)$", raw)
            if not m:
                return "Usage: set <text>"
            return ed.set(m.group(1))
        elif cmd == "clear":
            return ed.clear()
        elif cmd == "backspace" or cmd == "delete":
            n = 1 if len(parts) == 1 else int(parts[1])
            return ed.backspace(n)
        elif cmd == "insertat":
            m = re.match(r"^\s*insertat\s+(\d+)\s(.*)$", raw)
            if not m:
                return "Usage: insertat <pos> <text>"
            return ed.insert_at(int(m.group(1)), m.group(2))
        elif cmd == "replaceat":
            m = re.match(r"^\s*replaceat\s+(\d+)\s+(\d+)\s(.*)$", raw)
            if not m:
                return "Usage: replaceat <pos> <len> <text>"
            return ed.replace_range(int(m.group(1)), int(m.group(2)), m.group(3))
        elif cmd == "replaceall":
            m = re.match(r"^\s*replaceall\s+(\S+)\s(.*)$", raw)
            if not m:
                return "Usage: replaceall <old> <new>"
            return ed.replace_all(m.group(1), m.group(2))
        elif cmd == "undo":
            return ed.undo()
        elif cmd == "redo":
            return ed.redo()
        elif cmd == "exit":
            return "Bye!"
        else:
            return "Unknown command. Type 'help'."
    except Exception as e:
        return f"Error: {e}"


def run_commands(commands: List[str], ed: Optional[NotebookEditor] = None) -> List[str]:
    """Run a list of commands and return their outputs in order."""
    if ed is None:
        ed = NotebookEditor()
    outputs: List[str] = []
    for raw in commands:
        outputs.append(run_command(raw, ed))
    return outputs
```

## Interactive CLI
Execute the following cells to start an input-driven loop. Enter commands such as `insert 0 Hello`, `undo`, `redo`, `show`, and `exit`.

```python
# Interactive loop definition
def start_cli(ed: Optional[NotebookEditor] = None) -> None:
    """Start an interactive command loop. Stop with 'exit' or EOF (Ctrl+D)."""
    if ed is None:
        ed = NotebookEditor()
    print("Undo/Redo CLI")
    print("Type 'help' for commands.")
    while True:
        try:
            raw = input("> ")
        except EOFError:
            print("Bye!")
            break
        out = run_command(raw, ed)
        if out:
            print(out)
        if raw.strip().lower() == "exit":
            break
```

```python
# Start the interactive CLI (runs until you type `exit`)
start_cli()
```
